```python
import pandas as pd
file_path = r"/Users/cristinaortiz/Downloads/dognition_data_aggregated_by_dogid (2).xlsx"
data = pd.read_excel(file_path)

import matplotlib.pyplot as plt
dog_training_avg = data.groupby("Dog_Fixed")["Total Tests Completed"].mean()

plt.figure(figsize=(8, 5))
dog_training_avg.plot(kind="bar", color="skyblue")
plt.title("Impact of Behavioral Training on Tests Completed")
plt.xlabel("Behavioral Training (1 = Yes, 0 = No)")
plt.ylabel("Total Tests Completed")
plt.xticks([0, 1], labels=["No Training", "Training"])
plt.yticks(range(10, 25,5))
plt.show()
```
