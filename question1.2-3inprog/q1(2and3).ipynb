```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sqlite3
import os
```

```python
dognition_agg= pd.read_excel("dognition_data_aggregated_by_dogid.xlsx")
clean_dognition = dognition_agg.dropna()
#display(clean_dognition)
```

```python
#(Part 2)
sns.scatterplot(data=clean_dognition, x='Total Tests Completed', y='Time diff between first and last game (days)')
plt.xlabel("Total Number of Tests Completed")
plt.ylabel("Time Difference Between First & Last Game (Days)")

plt.title("Number of Total Tests And Speed of Game (in Days) Correlation")
```

```python
sns.scatterplot(data=clean_dognition, x='Total Tests Completed', y='Time diff between first and last game (minutes)')
plt.xlabel("Total Number of Tests Completed")
plt.ylabel("Time Difference Between First & Last Game (Minutes)")
plt.title("Number of Total Tests And Speed of Game (in Min) Correlation")
```

```python
#(PART 3)
#total number of tests for EACH dog breed
tests_each_breed= clean_dognition.groupby('Breed_Type')['Total Tests Completed'].sum()
print(tests_each_breed)

tests_each_breed.plot(x='Breed_Type', y='Total Tests Completed', kind='bar');
plt.xlabel('Breed')
plt.ylabel('Sum of Total Tests')
plt.title('Sum of Total Tests Completed for Each Dog Breed')
```

```python
#total number of tests for EACH dog group

tests_each_breed= clean_dognition.groupby('Breed_Group')['Total Tests Completed'].sum()
print(tests_each_breed)

tests_each_breed.plot(x='Breed_Group', y='Total Tests Completed', kind='bar');
plt.xlabel('Dog Group')
plt.ylabel('Sum of Total Tests')
plt.title('Sum of Total Tests Completed for Each Dog Group')
```
